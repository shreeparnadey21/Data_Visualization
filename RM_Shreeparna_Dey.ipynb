```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('reservoir-cwc-jan22_0.csv')
df.columns
```

```python
# df[df['State']=='Tamil Nadu'].shape
```

```python
df1=df[df['State']=='Tamil Nadu']
df2=df1[df1['Reservoir_name']=='Aliyar Reservoir']
df2
```

```python
plt.figure(figsize=(10,5))
plt.scatter(df2['Date'],df2['Storage'])
plt.xlabel('Date')
plt.ylabel('Storage')
plt.xticks(rotation=90)
plt.grid()
plt.title('Scatter plot of Storage vs Date for Aliyar Reservoir,Tamil Nadu in month of Jan 2022')
plt.show()
```

```python
df['State'].unique()
```

```python
df1=df[df['State']=='Karnataka']
rv=df1['Reservoir_name'].unique()
x=[]
y=[]
for i in rv:
    y.append(list(df1[df1['Reservoir_name']==i]['Storage']))
    
plt.figure(figsize=(15,8))
plt.boxplot(y)
plt.xlabel('Karnataka Reservoirs')
plt.ylabel('Storage')
plt.xticks(list(range(1,len(rv)+1)),list(rv),rotation=90)
plt.grid()
plt.title('Box plot of Storage accross different Reserviors in Karnataka in month of Jan 2022')
plt.show()
```

```python
df1=df[df['State']=='Gujarat']
rv=df1['Reservoir_name'].unique()
x=[]
y=[]
for i in rv:
    y.append(list(df1[df1['Reservoir_name']==i]['Full_reservoir_level'].unique())[0])
    
plt.figure(figsize=(15,8))
plt.bar(list(rv),y)
plt.xlabel('Gujarat Reservoirs')
plt.ylabel('Full_reservoir_level')
plt.xticks(rotation=90)
plt.grid()
plt.title('Bar plot of Full_reservoir_level accross different Reserviors in Gujarat in month of Jan 2022')
plt.show()
```
